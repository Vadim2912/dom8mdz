# Задание 1 
Получить интуицию, на что влияют параметры распределений. 
Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, 
Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
import random

%matplotlib inline
```

```python
num = 1

while num < 6:
    
    mu = random.uniform(2.0,4.0)
    sigma = random.uniform(0.1,0.6)
    print(mu, sigma)

    # зададим нормально распределенную случайную величину
    norm_rv = sts.norm(loc=mu, scale=sigma)

    # сгенерируем 10 значений
    norm_rv.rvs(size=10)

    x = np.linspace(0,6,100)
    pdf = norm_rv.pdf(x)
    plt.plot(x, pdf)
    
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    num += 1
    
plt.show()
```

```python
# Экспоненциальное
num = 1

while num < 6:
    
    lam = random.uniform(0.1,1.0)
    expon_rv = sts.expon(lam)
    expon_rv.rvs(10)

    e = np.linspace(0,5,1000)
    pdfe = expon_rv.pdf(e)

    plt.plot(e, pdfe)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    num += 1

plt.show()
```

```python
# Стьюдента
num = 1

while num < 6:
    
    n = random.uniform(20,100)
    t_rv = sts.t(n)
    t_rv.rvs(10)
    t = np.linspace(0,10,1000)
    pdft = t_rv.pdf(t-5)

    plt.plot(t, pdft)
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
    num += 1

plt.show()
```
